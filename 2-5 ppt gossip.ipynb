```python
import requests
from bs4 import BeautifulSoup

def get_articles(url):
    response = requests.get(url, cookies=cookies)
    soup = BeautifulSoup(response.text, 'lxml')

    for article in soup.find_all('div', class_='r-ent'):
        title_div = article.find('div', class_='title')
        if title_div.a:
            title = title_div.a.text
            href = 'https://www.ptt.cc' + title_div.a.get('href')
        else:
            title = title_div.text.strip()
            href = ' '

        date = article.find('div', class_='date').text
        print(date, href, title)

    #得到下一頁連結
    next_page = soup.find_all('a', class_='btn wide')[1].get('href')
    return 'https://www.ptt.cc' + next_page

# 主程式
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
cookies = {'over18': '1'}

for i in range(5):
    url = get_articles(url)
```
