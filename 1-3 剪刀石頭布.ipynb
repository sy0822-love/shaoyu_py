```python
import random

def play_game():
    choices = ["scissors", "rock","paper"]

    player = input("please enter your choice: ")

    if player not in choices:
        print("error！")
        return

    #random choices
    computer = random.choice(choices)

    #choices
    print(f"you: {player}")
    print(f"computer: {computer}")

    # win or lose
    if player == computer:
        result = "tie！"
    elif (player == "scissors" and computer == "paper") or \
         (player == "rock" and computer == "scissors") or \
         (player == "paper" and computer == "rock"):
        result = "you win"
    else:
        result = "you lose"

    print(result)

play_game()
```
