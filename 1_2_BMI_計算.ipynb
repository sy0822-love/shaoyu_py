```python
def calculate_bmi():
    try:
        weight = float(input("enter your weight(kg): "))
        height = float(input("enter your height (m): "))

        #  BMI calculation
        bmi = weight / (height ** 2)

        #  BMI classification
        if bmi < 18.5:
            category = "underwight"
        elif 18.5 <= bmi < 24:
            category = "normal"
        elif 24 <= bmi < 27:
            category = "overweight"
        elif 27 <= bmi < 30:
            category = "mild obesity"
        else:
            category = "moderate and severe obesity"

        # result output
        print(f"your bmi: {bmi:.2f}")
        print(f"your weight classification: {category}")

    except ValueError:
        print("errorï¼")

calculate_bmi()
```
